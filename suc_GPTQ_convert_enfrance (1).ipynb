```python
from transformers import AutoTokenizer, TextGenerationPipeline
from auto_gptq import AutoGPTQForCausalLM

# معرف النموذج المراد استخدامه (ضع هنا معرف النموذج المطلوب)
model_id = "rakmik/lama3.21b-gptq"  # استبدل بالمعرف الصحيح

# تحميل التوكنيزر
tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=True)

# تحميل النموذج الكمي
model = AutoGPTQForCausalLM.from_quantized(model_id, device="cuda:0")  # تشغيل على GPU

# استدلال باستخدام model.generate
prompt = "What is the capital of France?"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
output_ids = model.generate(**inputs)

# طباعة النتيجة
generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("Generated Text:", generated_text)

# الاستدلال باستخدام Pipeline
pipeline = TextGenerationPipeline(model=model, tokenizer=tokenizer)
print(pipeline(prompt)[0]["generated_text"])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
model_name = "rakmik/lama3.21b-gptq"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

prompt = "Give me a short introduction to large language model."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt},
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512,
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "rakmik/lama3.21b-gptq"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token  # Very important for some models
# Define the special tokens (check the model's documentation!)
BOS_TOKEN = "<s>"  # Beginning of sequence
EOS_TOKEN = "</s>"  # End of sequence
SYSTEM_TOKEN = "<|system|>"
USER_TOKEN = "<|user|>"
ASSISTANT_TOKEN = "<|assistant|>"

prompt = "Give me a short introduction to large language model."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt},
]

# Manually format the prompt
text = ""
for message in messages:
    if message["role"] == "system":
        text += SYSTEM_TOKEN + message["content"] + EOS_TOKEN
    elif message["role"] == "user":
        text += USER_TOKEN + message["content"] + EOS_TOKEN
    elif message["role"] == "assistant":
        text += ASSISTANT_TOKEN + message["content"] + EOS_TOKEN
text = BOS_TOKEN + text + ASSISTANT_TOKEN  # Add BOS and start the assistant

model_inputs = tokenizer([text], return_tensors="pt", padding=True).to(model.device)  #padding added.
#print(model_inputs)
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=11,
    pad_token_id=tokenizer.eos_token_id  # Important for generation!
)

generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```
